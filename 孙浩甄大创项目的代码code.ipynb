中文分词

```python
import xlwt
import jieba

#获得去除标点的文本
def get_text(file_name):
  with open(file_name, 'r', encoding='utf-8') as fr:
    text = fr.read()
    #要删除的标点
    del_ch = ['《','，','》','\n','。','、','；','"',\
      '：',',','！','？',' ']
    for ch in del_ch:
      text = text.replace(ch,'')#这里无需替换成空格
    return text

#文件名改为要分析的文件
file_name = '/content/贺丽霞.txt'
text = get_text(file_name)
vlist = jieba.lcut(text)#调用jieba实现分词，返回列表

res_dict = {}
exclude = ['却说','不能','如何','左右','二人','不可',\
  '荆州','如此','商议','主公','军士']



#进行词频统计
for i in vlist:
  if len(i)<=1:
    continue
  elif i in exclude:
    continue
  elif i == '建议说':
    ri = '建议'
  elif i == '休息':
    ri = '休息'
  elif i == '如何':
    ri = '如何'
  elif i == '如何':
    ri = '如何'
  elif i == '怎么样':
    ri = '怎么样'
  elif i == '联系我':
    ri = '联系我'    
     
  else:
    ri = i
  res_dict[ri] = res_dict.get(ri,0) + 1
res_list = list(res_dict.items())
#降序排序
res_list.sort(key = lambda x:x[1], reverse = True)
 

for i in range(500):
  word,count = res_list[i]
  pstr = str(i+1) + ':'
  print(pstr, end=' ')
  print(word,count)


#文件名改为要分析的文件
file_name = '/content/emotion.txt'
text2 = get_text(file_name)
vlist2 = jieba.lcut(text2)#调用jieba实现分词，返回列表

wb = xlwt.Workbook()
sheet = wb.add_sheet("第一个 sheet")
head_data = ['word','count']
for head in head_data:
	sheet.write(0, head_data.index(head), head)
# sheet.write(x行, y列, head)

for i in range(1,500):
  word,count=res_list[i]  
  emotion=0
  for j in range(1,26):
    if word==vlist2[j]:
      emotion=1
  sheet.write(i, 0, word) # row, column, value
  sheet.write(i, 1, count) # row, column, value
  sheet.write(i, 2, emotion)# row, column, value
wb.save('贺丽霞结果.xls')








```
